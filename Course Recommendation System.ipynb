```python
data = data[['Course Name', 'Difficulty Level', 'Course URL', 'Course Description', 'Skills']]
```

```python
data.head()
```

```python
# Clean 'Course Name' column
data.loc[:, 'Course Name'] = (
    data['Course Name']
        .str.replace(':', '', regex=False)
        .str.replace(' ', ',', regex=False)
        .str.replace(',,', ',', regex=False)
)

# Clean 'Course Description' column
data.loc[:, 'Course Description'] = (
    data['Course Description']
        .str.replace(':', '', regex=False)
        .str.replace('_', '', regex=False)
        .str.replace('[()]', '', regex=True)
        .str.replace(' ', ',', regex=False)
        .str.replace(',,', ',', regex=False)
)

# Clean 'Skills' column
data.loc[:, 'Skills'] = data['Skills'].str.replace('[()]', '', regex=True)
```

```python
data.head()
```

```python
data['tags'] = data['Course Name'] + data['Difficulty Level'] + data['Course Description'] + data['Skills']
```

```python
data.head()
```

```python
data['tags'].iloc[1]
```

```python
new_df=data[['Course Name','Course URL','tags']]
```

```python
new_df["Course Name"] = data['Course Name'].str.replace(',', ' ')
```

```python
new_df["Course Name"] 
```

```python
new_df.rename(columns = {"Course Name" : "course_name", "Course URL" : "course_url"}, inplace = True)
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
new_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words = 'english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors[0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(course):
    course_index = new_df[new_df['course_name'] == course].index[0]
    distance = similarity[course_index]
    course_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x:x[1])[1:7]
    
    recommended_courses = []
    for i in course_list:
        course_name = new_df.iloc[i[0]].course_name
        course_url = new_df.iloc[i[0]].course_url
        recommended_courses.append({'name': course_name, 'url': course_url})
    return recommended_courses        
```

```python
recommend("Retrieve Data using Single-Table SQL Queries")
```

```python
import pickle
from pathlib import Path

# Use your specified path
base_path = Path(r"G:\Protfolio Projects\Course Recommendation System\models")

# Create the directory if it doesn't exist
base_path.mkdir(parents=True, exist_ok=True)

# Save similarity matrix
with open(base_path / "similarity.pkl", "wb") as f:
    pickle.dump(similarity, f)

# Save course list as dictionary records
course_list = new_df[['course_name', 'course_url']].to_dict('records')
with open(base_path / "course_list.pkl", "wb") as f:
    pickle.dump(course_list, f)

# Save course DataFrame directly
courses = new_df[['course_name', 'course_url']]
with open(base_path / "courses.pkl", "wb") as f:
    pickle.dump(courses, f)
```
